```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../../../training_results.csv")
```

```python
df.head()
```

```python
plt.plot(df["loss"], label="Train loss")
plt.plot(df["loss_val"], label="Val loss")
plt.title("Biểu đồ thể hiện độ lỗi log của train và val")
plt.legend()
```

```python
plt.plot(df["mrr"], label="Train MRR")
plt.plot(df["mrr_val"], label="Val MRR")
plt.title("Biểu đồ thể hiện độ đo MRR của train và val")
plt.legend()
```
